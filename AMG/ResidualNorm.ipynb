```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Datos(path):
    data = []
    with open(path) as fobj: 
        for i, line in enumerate(fobj):
            if line.rsplit() != []:
                row = line.rsplit()   
                data.append(float(row[0]))     
    return np.array(data)
```

```python
Nx, Nt = 16, 16
m0 = -0.7
fig = plt.figure(dpi=150)
plt.title(r'Residual norm of the coarse level solution.'+ '\n'+ '$N_x$={0}, $N_t={1}$, $m_0$={2}'.format(Nx,Nt,m0))
plt.ylabel(r'$||r||$')
plt.xlabel("Iteration")

it = np.arange(1,51,1)

path = 'build/r_norm_m{0}.txt'.format(str(m0).replace("-","").replace(".",""))
#path = 'build/r_norm_m075_1.txt'

residual = Datos(path)
plt.plot(it[:20],residual[:20],linestyle='',marker='x',markersize=5) 
#plt.legend()
plt.xticks(np.arange(1,21,2))
plt.tight_layout()
plt.show()

fig.savefig("r_norm_m{0}.pdf".format(str(m0).replace("-","").replace(".","")))
for r in residual:
    print(r)
```
